```python
import re
import pandas as pd
sms = pd.read_csv('spam.csv',encoding='ISO-8859-1');
sms.head()
#nltk.download('stopwords')
#nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords 

#Preprocessing the data: Remove none alphabetic characters/Make the word lower case/Remove the stop words

def mutlti_func(x):       
    WT_sms= word_tokenize(str(re.sub('[^A-Za-z]', ' ', x)).lower())
    for word in WT_sms:
        if word in stopwords.words('english'):
            WT_sms.remove(word)
    return WT_sms


sms['v2']=sms['v2'].apply(mutlti_func)
sms.head()
```

```python
from nltk.stem.porter import PorterStemmer
from spellchecker import SpellChecker


# Stemming / Spelling 
spell = SpellChecker()
stemmer = PorterStemmer()

def stem_spell_fuc(x):
    for i in range(len(x)):
        #x[i] = stemmer.stem(spell.correction(x[i]))
        x[i] = stemmer.stem(x[i])
    return x

sms['v2']=sms['v2'].apply(stem_spell_fuc)
sms.head()
```

```python
def join_fuc(x):
    x = " ".join(x)
    return x

sms.columns=["v1","v2","v3","v4","v5"]
sms['v3']=sms['v2'].apply(join_fuc)
sms.head()
```

```python
def joincomm(x):
    temp2=re.sub(" ",",",x)
    return temp2
sms['v4']=sms['v3'].apply(joincomm)
sms.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(token_pattern=u'(?u)\\b\\w+\\b')
vect = CountVectorizer(token_pattern=u'(?u)\\b\\w+\\b')
result=vect.fit_transform(sms['v3']).toarray()

result
```
